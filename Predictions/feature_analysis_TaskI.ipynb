```python
import re
#convert to bandage labels
def get_products(dat):
    dat = dat.values
    bio_names = list()
    for e in dat:
        if 'Name=' in e:
            try:
                lbl = re.findall(r'Name=([\w\)\(-]+);', e)[0]
                if re.match(r'([a-zA-Z0-9]+)_[0-9]+', lbl):
                    lbl = re.findall(r'([a-zA-Z0-9]+)_[0-9]+', lbl)[0]
            except IndexError:
                print(e)
        elif 'eC_number=' in e:
            try:
                lbl = re.findall(r'eC_number=([0-9\.-]+);',e)[0]
            except IndexError:
                print(e)
        elif 'product=' in e:
            try:
                lbl = re.findall(r'product=([\w ]+)', e)[0]
            except IndexError:
                print(e)
        elif 'CRISPR' in e:
            lbl = 'CRISPR'
        bio_names.append(lbl)
    return set(bio_names)


for gff in glob.glob('/burg/pmg/users/ic2465/Projects/MANU_copangraph/data/Predictions/plt/TaskI/edgesync2_features_greyscale/*subgraph.gff'):
    df = pd.read_csv(gff, sep='\t', header=None)
    labels = df.groupby(3).apply(lambda x: pd.Series([x.iloc[0, 3], ', '.join(get_products(x.loc[:,8]))], index=['Node', 'product']))
    labels.to_csv(gff.replace('.gff', 'labels.csv'), index=None)
```

```python
import matplotlib.pyplot as plt


def blend_red_blue_to_gray(a, b, max_val):
    if a + b == 0:
        return '#808080'
    total = a + b
    red_ratio = a / total
    blue_ratio = b / total
    r = red_ratio * 255
    g = 0
    b_val = blue_ratio * 255
    gray = 128
    grayness = 1 - min(total / max_val, 1)
    r = r + (gray - r) * grayness
    g = g + (gray - g) * grayness
    b_val = b_val + (gray - b_val) * grayness
    return '#{0:02x}{1:02x}{2:02x}'.format(int(r), int(g), int(b_val))
def plot_lower_triangle_grid(p, c):
    fig, ax = plt.subplots(figsize=(p * 0.3, c * 0.3))
    max_val = p+c
    

    for a in range(p + 1):
        for b in range(c + 1):
            color = blend_red_blue_to_gray(a, b, max_val)
            # Position (b, a) so that x = b, y = a; (0,0) is lower-left
            rect = plt.Rectangle((b, a), 1, 1, facecolor=color, edgecolor='black', linewidth=0.3)
            ax.add_patch(rect)

    # Set axis limits and ticks
    ax.set_xlim(0, c + 1)
    ax.set_ylim(0, p + 1)
    #ax.set_xticks(np.array(range(c + 1)) + 0.5)
    #ax.set_yticks(np.array(range(p + 1)) + 0.5)
    #ax.set_xticklabels(range(c+1))
    #ax.set_yticklabels(range(p+1))
    ax.set_aspect('equal')
    plt.grid(False)
    plt.tight_layout()
    plt.show()
plot_lower_triangle_grid(36, 24)
```
